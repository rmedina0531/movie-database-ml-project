```python
#This Cell is just to import all the necessary things

#import KNeighborsClassifier "Class"
from sklearn.neighbors import KNeighborsClassifier

# Importing the required packages and libraries
import numpy as np
import pandas as pd

# Imports the load_iris function from sklearn library. 
from sklearn.datasets import load_iris 
```

```python
#A :Read the Kobe Shot Selection dataset and assign it to a Pandas DataFrame
# reading a CSV file, and store it in a pandas DataFrame:
kobe_df = pd.read_csv('/Users/Work/Documents/Coding/DataScience/Project/data.csv')
kobe_df
```

```python
#B: Split the dataset into testing and training sets with the following parameters: test_size=0.4, random_state=6

# create a python list of feature names that would like to pick from the dataset:
feature_cols = ['combined_shot_type','shot_type','shot_zone_range','minutes_remaining']

#get the features from the data set
X = kobe_df[feature_cols] 

#get the labels/targets
y = kobe_df['shot_made_flag']

# Randomly splitting the original dataset into training set and testing set
# The function"train_test_split" from "sklearn.cross_validation" library performs random splitting.

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=6)
```

```python
#C: Instantiate a KNN object with K=3, train it on the training set and test it on the testing set. 
    #Then, calculate the accuracy of your prediction as you learned in Lab3.

# "knn" is instantiated as an "object" of KNeighborsClassifier "class". 

K = 3
knn = KNeighborsClassifier(n_neighbors=K)

#Train the Model with training set
knn.fit(X_train, y_train)

#Test on the testing set using predict
y_predict = knn.predict(X_test)

print(y_predict)
```

```python
#C continued
#Testing the Accuracy
#import to calculate the accuracy
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_predict)

print(accuracy)
```

```python
#D Repeat part (c) for K=1, K=5, K=7, K=11, K=15, K=27, K=59 (you can simply use a “for loop,” and save the final 
    #accuracy results in a list). Does the accuracy always get better by increasing the value K?
listAccuracy = []
Kvalues = [1, 5, 7, 11,15,37,59]

for x in Kvalues:
    K = x
    
    knn = KNeighborsClassifier(n_neighbors=K)

    #Train the Model with training set
    knn.fit(X_train, y_train)

    #Test on the testing set using predict
    y_predict = knn.predict(X_test)
    
    #Testing the Accuracy
    #import to calculate the accuracy
    from sklearn.metrics import accuracy_score

    accuracy = accuracy_score(y_test, y_predict)
    listAccuracy.append(accuracy)

    print('Accuracy for value', K,' is ', accuracy)
    
print(listAccuracy)

```

```python
print('The accuracy does not get better as you increase K')
```
