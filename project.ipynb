```python
from ast import literal_eval

import numpy as np
import pandas as pd

from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

```

```python
credits_df = pd.read_csv('data/credits.csv')
movies_df = pd.read_csv('data/movies_metadata.csv')
```

```python
credits_df.head()
```

```python
movies_df.head()
```

```python


# print(movies_df['id'])
movies_df['id'] = pd.to_numeric(movies_df['id'])
df = movies_df.merge(credits_df, on='id')
df.head()

# print(type(credits_df['id'][0]))
# print(type(movies_df['id'][0]))
```

```python
df.set_index('title', inplace=True)
df
```

```python
df_reduced = df[['budget', 'genres', 'cast', 'crew', 'revenue', 'production_companies']]
df_reduced
```

# Getting the director Data

```python
df_reduced['crew'] = df_reduced['crew'].apply(literal_eval)
df_reduced['crew']
```

```python
# df_reduced['crew'][0][0]['name']
type(df_reduced['crew'][0])
```

```python
#data[0][0]['name']
# data
df_reduced['crew'] = df_reduced['crew'].apply(lambda x: [i['name'] for i in x] if isinstance(x, list) else [])
df_reduced['crew']
```

```python
#remove entries that have not crew data
df_reduced = df_reduced[df_reduced['crew'].map(lambda d: len(d)) > 0]

#get first entry of the crew list for director name
df_reduced['director'] = df_reduced['crew'].apply(lambda x: x[0])
df_reduced['director']
```

```python
df_reduced = df_reduced[['budget','genres','cast','revenue','production_companies','director']]
```

```python
df_reduced
```

# Cast Data

```python
df_reduced['cast'] = df_reduced['cast'].apply(literal_eval)
```
