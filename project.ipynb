### reducing toal number of actors, production companies, and directors to evaluate


```python
from sklearn.preprocessing import MultiLabelBinarizer
mlb = MultiLabelBinarizer()

def counter(column):
    labels = mlb.fit_transform(column)
    label_classes = mlb.classes_
    
    label_data = pd.DataFrame(labels, columns=label_classes)
    
    val = {}
    for x in label_classes:
        val.update({x:label_data[x].value_counts()[1]})
        
    sorted_val = sorted(val.items(), key=lambda kv: kv[1], reverse=True)
    val_pd = pd.DataFrame.from_dict(sorted_val, orient='columns')
    val_pd.rename(columns={0: "Value", 1: "count"}, inplace=True)
    return val_pd
```

### Using budget and director to determine revenue

```python
features = df_reduced[['budget','director']]
labels = df_reduced['revenue']
#using a label encoder on directors due to large amount of possible values
from sklearn.preprocessing import LabelEncoder

labelencoder = LabelEncoder()

features['director'] = labelencoder.fit_transform(features[['director']])
features
```

```python
#scaling the budget Data
from sklearn.preprocessing import scale

features['budget'] = scale(features['budget'])
features

```

```python
#create training and testing data
from sklearn.model_selection import train_test_split


#find out what values to use for test_size and random_state
feature_train, feature_test, label_train, label_test = train_test_split(
    features, labels, test_size=0.25, random_state=6)
```

```python
feature_train.shape
```

```python
feature_test.shape
```

```python
label_train.shape
```

```python
label_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

my_logreg = LogisticRegression()
#random state can be deleted
my_logreg.fit(feature_train, label_train)
y_predict_lr = my_logreg.predict(feature_test)
#accuracy score for Logistic Regression
score_lr = accuracy_score(label_test, y_predict_lr)


print("Accuracy for Logistic Regression:")
print(score_lr)
```
